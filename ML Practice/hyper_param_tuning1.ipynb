```python
### Author : Prasad Meesala
# Importing the necessary modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import ensemble, model_selection, metrics
```

```python
# Collecting the data

df = pd.read_csv(r"C:\Users\meesa\Desktop\VSC-ML\Datasets\voice.csv")
df.info()
df.head()
```

```python
# Data preprocessing

gender = pd.get_dummies(df['gender'])
df = pd.concat([df, gender], axis = 1)
df['target'] = df['female']
df.drop(['male', 'gender', 'female'], axis = 1, inplace = True)
df.head()
```

```python
# Splitting the data into training and testing data

X = df.drop(['target'], axis = 1)
y = df['target']

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
# Hyperparamter tuning

param_grid = [
    {'n_estimators' : [20, 50, 100], 'criterion' : ['gini', 'entropy', 'log_loss']},
]

rf = ensemble.RandomForestClassifier()

grid_search = model_selection.GridSearchCV(rf, param_grid = param_grid)

grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_)
print(grid_search.best_estimator_)
```

```python
# Training the model

model = ensemble.RandomForestClassifier(n_estimators = 100, criterion = 'log_loss')
model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```
